# Outlier Detection

```python
df_stu = df.copy()
```

```python
df_stu
```

```python
def outlier_detection(columns):
    outlier_indices = []
    for col in columns:
        Q1 = df_stu[col].quantile(0.25)
        Q3 = df_stu[col].quantile(0.75)
        IQR = Q3-Q1

        mask = (df_stu[col] >= (Q1-1.5*IQR))&(df_stu[col] <= (Q3+1.5*IQR))
        mask = mask.to_numpy()

        false_indices = np.argwhere(~mask)
        outlier_indices.append(false_indices)
    return np.unique(np.concatenate(outlier_indices).ravel())
```

## .ravel>> flattens the multi-dimensional array into 1 d array
        

```python
num_cols = ['age','G3']
```

# Target Variable is G3

```python
outlier_detection(num_cols)
```

```python
x, y = df_stu.iloc[:,:-1],df_stu.iloc[:,-1]
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2,random_state = 42)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
x_train.dtypes
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for col in x_train.select_dtypes(include = 'str').columns:
    x_train[col] = le.fit_transform(x_train[col])
    x_test[col] = le.fit_transform(x_test[col])
```

```python
y.unique()
```

```python
y.dtypes
```

```python
y
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error,r2_score
print("MAE : ",mean_absolute_error(y_test,y_pred))
print("R2 : ",r2_score(y_test,y_pred))
```

# Actual vs Predicted Grades

```python
error = np.abs(y_test - y_pred)

plt.figure(figsize=(8,6))

scatter = plt.scatter(y_test,y_pred,c=error,cmap='icefire',s=80,alpha=0.8)
plt.plot([y_test.min(), y_test.max()],[y_test.min(), y_test.max()],'--',color='black',linewidth=2)
plt.colorbar(scatter, label='Prediction Error')

plt.xlabel("Actual G3")
plt.ylabel("Predicted G3")
plt.title("Actual vs Predicted Grades")
plt.grid(alpha=0.3)


plt.show()
```

# Feature Importance

### Which factors affect final grades the most?

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(random_state=42)
rf.fit(x_train, y_train)

importance = pd.DataFrame({'Feature': x_train.columns,'Importance': rf.feature_importances_})
importance = importance.sort_values(by='Importance',ascending=False)
importance.head(10)
```

```python
plt.figure(figsize=(10,6))

sns.barplot(data=importance.head(10), x='Importance',y='Feature',palette='crest')
plt.title('Top 10 Important Features')
plt.show()
```
